```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import pearsonr, spearmanr
from itertools import combinations

plt.style.use('seaborn-v0_8')
sns.set_palette("Set1")
```

```python
# Загружаем датасет прямо из интернета
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)
```

```python
print("===ОСНОВНАЯ СТАТИСТИКА ПО ДАТАСЕТУ===")
print(f"Размерность датасета: {df.shape}")
print(f"Количество признаков: {df.shape[1]}")
print()

print("Количество объектов в каждом классе:")
print(df['species'].value_counts())
print(f"Количество целевых классов: {df['species'].nunique()}")
print()

# Процент объектов (строк) с хотя бы одним пропущенным значением
rows_with_missing = df.isnull().any(axis=1).sum()
total_rows = df.shape[0]
missing_percent = (rows_with_missing / total_rows) * 100
print(f"Процент объектов с неопределенными признаками: {missing_percent:.2f}%")
print(f"Количество строк с пропущенными значениями: {rows_with_missing} из {total_rows}")
print()

print("Количество пропущенных значений по каждому признаку:")
print(df.isnull().sum())
print()

print("Основные статистические характеристики:")
print(df.describe())
print()

print("Типы данных признаков:")
print(df.dtypes)
print()
```

```python
# Выбираем три класса и четыре количественных признака
selected_features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
selected_species = ['Adelie', 'Chinstrap', 'Gentoo']  # выбираем три класса

print("=== ФОРМИРОВАНИЕ ОТФИЛЬТРОВАННОГО ДАТАСЕТА ===")
print(f"Выбранные количественные признаки: {selected_features}")
print(f"Выбранные классы: {selected_species}")
print()

df_filtered = df[df['species'].isin(selected_species)].copy()
df_filtered = df_filtered[['species'] + selected_features].dropna()

print("Статистика после фильтрации:")
print(f"Размерность отфильтрованного датасета: {df_filtered.shape}")
print(f"Количество признаков: {df_filtered.shape[1]}")
print()

print("Количество объектов в каждом классе после фильтрации:")
print(df_filtered['species'].value_counts())
print()

print("Пропущенные значения в выбранных признаках после фильтрации:")
print(df_filtered[selected_features].isnull().sum())
print()

print("Первые 5 строк отфильтрованного датасета:")
print(df_filtered.head())
```

# 3 ПУНКТ

```python
feature_pairs = list(combinations(selected_features, 2))

species_colors = {'Adelie': 'red', 'Chinstrap': 'blue', 'Gentoo': 'green'}
species_markers = {'Adelie': 'o', 'Chinstrap': 's', 'Gentoo': '^'}

fig, axes = plt.subplots(2,3, figsize=(18,12))
fig.suptitle('Pairwise Scatter Plots по парам количественных переменных', fontsize=16)

axes_flat = axes.flatten()

for idx, (x_feature, y_feature) in enumerate(feature_pairs):
    ax = axes_flat[idx]
    
    # Рисуем точки для каждого класса
    for species in selected_species:
        species_data = df_filtered[df_filtered['species'] == species]
        ax.scatter(species_data[x_feature], species_data[y_feature],
                  c=species_colors[species], marker=species_markers[species],
                  label=species, alpha=0.7, s=60)
    
    ax.set_xlabel(x_feature)
    ax.set_ylabel(y_feature)
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(4, 2, figsize=(15,16))
fig.suptitle('Гистограммы и плотности распределения по классам', fontsize=16)

for idx, feature in enumerate(selected_features):
    ax_hist = axes[idx, 0]
    for species in selected_species:
        species_data = df_filtered[df_filtered['species'] == species][feature]
        ax_hist.hist(species_data, bins=25, alpha=0.6, label=species,
                     color=species_colors[species], edgecolor='black', linewidth=0.3)

    ax_hist.set_xlabel(feature)
    ax_hist.set_ylabel('Частота')
    ax_hist.set_title(f'Гистограмма: {feature}')
    ax_hist.legend()
    ax_hist.grid(True, alpha=0.4)

    # Плотности
    ax_kde = axes[idx, 1]
    for species in selected_species:
        species_data = df_filtered[df_filtered['species'] == species][feature]
        sns.kdeplot(data=species_data, ax=ax_kde, label=species, 
                    color=species_colors[species], linewidth=2)
        ax_kde.set_xlabel(feature)
        ax_kde.set_ylabel('Плотность')
        ax_kde.set_title(f'Плотность: {feature}')
        ax_kde.legend()
        ax_kde.grid(True, alpha=0.4)

plt.tight_layout()
plt.show()
```

# 4 ПУНКТ

```python
def correlation_table(data, method_name, method_func):
    n_features = len(selected_features)
    corr_matrix = np.zeros((n_features, n_features))
    p_matrix = np.zeros((n_features, n_features))

    for i, feat1 in enumerate(selected_features):
        for j, feat2 in enumerate(selected_features):
            if i == j:
                corr_matrix[i, j] = 1.0
                p_matrix[i, j] = 0.0
            elif i < j:
                corr, p_value = method_func(data[feat1], data[feat2])
                corr_matrix[i, j] = corr_matrix[j, i] = corr
                p_matrix[i, j] = p_matrix[j, i] = p_value

            else:
                corr_matrix[i, j] = corr_matrix[j, i]
                p_matrix[i, j] = p_matrix[j, i]

    corr_df = pd.DataFrame(corr_matrix,
                           index=selected_features,
                           columns=selected_features) 
    
    p_df = pd.DataFrame(p_matrix,
                        index=selected_features,
                        columns=selected_features)

    return corr_df, p_df

print("Корреляции Пирсона для всего датасета:")
pearson_corr, pearson_p = correlation_table(df_filtered, "Pearson", pearsonr)
print(pearson_corr.round(4))
print("\nP-values для корреляций Пирсона:")
print(pearson_p.round(4))

print("\nКорреляции Спирмана для всего датасета:")
spearman_corr, spearman_p = correlation_table(df_filtered, "Spearman", spearmanr)
print(spearman_corr.round(4))
print("\nP-values для корреляций Спирмана:")
print(spearman_p.round(4))

# 2. Тепловые карты для всего датасета
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Тепловые карты корреляций для всего датасета', fontsize=16)

# Тепловая карта Пирсона
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', center=0, 
            square=True, ax=axes[0,0], fmt='.3f')
axes[0,0].set_title('Корреляции Пирсона')

# Тепловая карта Пирсона (p-values)
sns.heatmap(pearson_p, annot=True, cmap='viridis', 
            square=True, ax=axes[0,1], fmt='.3f')
axes[0,1].set_title('P-values Пирсона')

# Тепловая карта Спирмана
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', center=0, 
            square=True, ax=axes[1,0], fmt='.3f')
axes[1,0].set_title('Корреляции Спирмана')

# Тепловая карта Спирмана (p-values)
sns.heatmap(spearman_p, annot=True, cmap='viridis', 
            square=True, ax=axes[1,1], fmt='.3f')
axes[1,1].set_title('P-values Спирмана')

plt.tight_layout()
plt.show()
```

```python
# 3. Корреляции для каждого класса отдельно
for species in selected_species:
    print(f"\n=== КОРРЕЛЯЦИИ ДЛЯ КЛАССА: {species} ===")
    species_data = df_filtered[df_filtered['species'] == species]
    
    print(f"Корреляции Пирсона для {species} (n={len(species_data)}):")
    pearson_corr_species, pearson_p_species = correlation_table(species_data, "Pearson", pearsonr)
    print(pearson_corr_species.round(4))
    
    print(f"\nКорреляции Спирмана для {species} (n={len(species_data)}):")
    spearman_corr_species, spearman_p_species = correlation_table(species_data, "Spearman", spearmanr)
    print(spearman_corr_species.round(4))

# 4. Тепловые карты для каждого класса
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
fig.suptitle('Тепловые карты корреляций Пирсона по классам', fontsize=16)

for idx, species in enumerate(selected_species):
    species_data = df_filtered[df_filtered['species'] == species]
    pearson_corr_species, _ = correlation_table(species_data, "Pearson", pearsonr)
    
    sns.heatmap(pearson_corr_species, annot=True, cmap='coolwarm', center=0, 
                square=True, ax=axes[idx, 0], fmt='.3f', vmin=-1, vmax=1)
    axes[idx, 0].set_title(f'Пирсон: {species} (n={len(species_data)})')
    
    spearman_corr_species, _ = correlation_table(species_data, "Spearman", spearmanr)
    
    sns.heatmap(spearman_corr_species, annot=True, cmap='coolwarm', center=0, 
                square=True, ax=axes[idx, 1], fmt='.3f', vmin=-1, vmax=1)
    axes[idx, 1].set_title(f'Спирман: {species} (n={len(species_data)})')

plt.tight_layout()
plt.show()

# 5. Сводная таблица значимых корреляций (p < 0.05)
print("\n=== СВОДНАЯ ТАБЛИЦА ЗНАЧИМЫХ КОРРЕЛЯЦИЙ (p < 0.05) ===")

print("\nЗначимые корреляции Пирсона для всего датасета:")
significant_pearson = pearson_corr.abs()[pearson_p < 0.05]
significant_pearson = significant_pearson.where(significant_pearson != 1.0, 0)  # Убираем диагональ
print(significant_pearson.round(4))

print("\nЗначимые корреляции Спирмана для всего датасета:")
significant_spearman = spearman_corr.abs()[spearman_p < 0.05]
significant_spearman = significant_spearman.where(significant_spearman != 1.0, 0)  # Убираем диагональ
print(significant_spearman.round(4))
```
